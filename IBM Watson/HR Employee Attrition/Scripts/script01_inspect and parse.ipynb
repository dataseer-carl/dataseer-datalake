# Issues discovered

* *DailyRate*, *HourlyRate*, *MonthlyRate*, and *MonthlyIncome* redundant columns?
* Dummy columns: columns that only contain a single value
* Metadata

# Fix issues

```r
hr.df <- hr.raw
```

## Ambiguous *Rate* and *Income* columns

```r
GGally::ggpairs(hr.df %>% select(ends_with("Rate"), ends_with("Income"))) + theme_basic()
```

Plot above shows couple of things:

1. `HourlyRate`, `DailyRate`, and `MonthlyRate` are not related with one another.
1. Neither are they related to `MonthlyIncome`.
1. They are also uniformly distributed.
1. `MonthlyIncome` has a skewed distribution.

**Resolution**: Use `MonthlyIncome`; ditch everything else.

```r
hr.df %<>% select(-ends_with("Rate"))
str(hr.df)
```

## Dummy columns

The columns `EmployeeCount`, `Over18`, and `StandardHours` all have only one unique value.

**Resolution**: remove these columns

```r
hr.df %<>% select(-EmployeeCount, -Over18, -StandardHours)
str(hr.df)
```

## Metadata

Fix the ff:

* Logicals
* Percentages
* Factors
* Coded factors
* others

```r
hr.df %<>% 
    mutate(
        # Logicals
        Separated = Attrition == "Yes",
        OverTime = OverTime == "Yes",
        # Percent
        SalaryHike = PercentSalaryHike / 100,
        # Factors
        Department = as.factor(Department),
        Sex = as.factor(Gender),
        MaritalStatus = as.factor(MaritalStatus),
        StockOptionLevel = as.factor(StockOptionLevel),
        # Coded factors
        BusinessTravel = BusinessTravel %>% 
            str_replace_all(c("^Travel_" = "")) %>% 
            factor(c("Non-Travel", "Rarely", "Frequently")),
        Education = factor(
            Education, levels = 1:5, labels = c("Below College", "College", "Bachelor", "Master", "Doctor")
        ),
        EnvironmentSatisfaction = factor(
            EnvironmentSatisfaction, levels = 1:4, labels = c("Low", "Medium", "High", "Very high")
        ),
        JobInvolvement = factor(
            JobInvolvement, levels = 1:4, labels = c("Low", "Medium", "High", "Very high")
        ),
        JobSatisfaction = factor(
            JobSatisfaction, levels = 1:4, labels = c("Low", "Medium", "High", "Very high")
        ),
        PerformanceRating = factor(
            PerformanceRating, levels = 1:4, labels = c("Low", "Good", "Excellent", "Outstanding")
        ),
        RelationshipSatisfaction = factor(
            RelationshipSatisfaction, levels = 1:4, labels = c("Low", "Medium", "High", "Very high")
        ),
        WorkLifeBalance = factor(
            WorkLifeBalance, levels = 1:4, labels = c("Bad", "Good", "Better", "Best")
        ),
        # Prettify
        EmployeeNumber = str_pad(
            EmployeeNumber,
            max(str_length(EmployeeNumber)),
            side = "left", pad = "0"
        )
    ) %>% 
    select(-Attrition, -Gender, -PercentSalaryHike)
str(hr.df)
```
