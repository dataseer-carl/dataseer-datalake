### Create customer ID

```r
library(stringr)

test <- c("ROBERTSONLILLITH", "PERKINSTHOR", "GARCIAJEREMY", "MATHISBLAKE")
test.out <- str_split(test, "")
test.out <- lapply(test.out, match, table = LETTERS)
test.out <- sapply(test.out, paste, collapse = "")
str(test.out)

name2serial <- function(x){
    require("stringr")
    
    x.letters <- str_split(x, "")
    x.pos <- lapply(x.letters, match, table = LETTERS)
    x.id <- sapply(x.pos, paste, collapse = "")
    return(x.id)
}
```

```r
cust.ref <- lookup.ls$Item %>%
    distinct(Buyer) %>%
    mutate(
        String = Buyer %>% str_to_upper() %>% str_replace_all(c(", " = "")),
        CustomerID = name2serial(String)
    ) %>%
    select(-String)
str(cust.ref)
```

```r
store.district <- lookup.ls$Store %>%
    distinct(DistrictID, DistrictName, DM, DM_Pic) %>%
    arrange(DistrictID)
store.district
```

District columns in store lookup and district lookup are identical.

Thus, remove these columns from store lookup.

```r
lookup.ls$Store %<>% select(-DM, -DM_Pic)
str(lookup.ls$Store)
```

### Fix store lookup

```r
# Check uniqueness of LocationID and StoreNumber
lookup.ls$Store %>% distinct(LocationID, StoreNumber) %>% 
    mutate(isSame = LocationID == StoreNumber) %>%
    summarise(all(isSame))
```

`LocationID` and `StoreNumber` are identical.  Drop `StoreNumber`.

```r
# Check uniqueness of Open Month No and Open Month
lookup.ls$Store %>% distinct(`Open Month No`, `Open Month`) %>% arrange(`Open Month No`)
```

`Open Month No` and `Open Month` are identical.

```r
store.df <- lookup.ls$Store %>%
    mutate(
        LocationID = as.character(LocationID),
        District = DistrictName %>% str_extract("[[:digit:]]+$") %>% 
            str_pad(2, "left", "0") %>% paste("District", .),
        OpenDate = as.Date(OpenDate, origin = "1899-12-30")
    ) %>%
    select(
        LocationID, Name, Chain, District,
        `City Name`, Territory, PostalCode,
        OpenDate, `Store Type`, SellingAreaSize
    )
(store.df)
```

```r
district.df <- lookup.ls$District %>%
    select(-DM_Pic, -DMImage) %>% # Broken link
    mutate(
        Chain = case_when(
            str_extract(District, "^[[:alpha:]]{2}") == "FD" ~ "Fashions Direct",
            str_extract(District, "^[[:alpha:]]{2}") == "LI" ~ "Lindseys"
        ),
        District = District %>% str_extract("[[:digit:]]+$") %>%
            str_pad(2, "left", "0") %>% paste("District", .)
    ) %>%
    select(Chain, District, DM)
district.df
```
